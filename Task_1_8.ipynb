# Complex Conditions of Cycle Ending
Create an algorithm and a program

```python
#1
def localmax(f,x1,x2,dx,var):
    print('5 iteration steps'); t=[[]]
    for x in [x1,x1+dx,..,x2]:
        x0=x; t+=[[f(x),f(x+dx)]]
        if f(x+dx)<f(x): break
    show(table([['f(x)','f(x+dx)']]+t[-5:]))
    (plot(f,(x1,x2))+point((x0,f(x0)),color='red')).show(gridlines=True,figsize=(7,4))
    print('var %d: f(x)=max => x≈%.10f'%(var,x0))
X1,X2,dX=.1,10,.1; f1(x)=(x+1)/(x^5+1)+log(x+1)
localmax(f1,X1,X2,dX,1)
```

```python
#2
def nulfunction(f,x1,x2,dx,var):
    print('iteration steps'); t=[['f(x)','f(x+dx)']]; x=x1;    
    while sign(f(x))==sign(f(x+dx)):
        x+=dx; t+=[[f(x),f(x+dx)]]
    show(table(t))
    (plot(f,(x1,x2))+point((x+dx,f(x+dx)),color='red')).show(gridlines=True,figsize=(7,4))
    print('var %d: f(x)=0 => x≈%.1f (solve: %.10f)'%(var,x+dx,find_root(f,x1,x2)))
X1,X2,dX=.1,10,.1; f2(x)=x-atan(sqrt(x)); nulfunction(f2,X1,X2,dX,2)
```
