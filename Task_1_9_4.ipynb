# Determine the Array Points Which Fall in a Given Area
Create an algorithm and a program

```python
#1
var('x,y')
[cx1,cx2,cx3,cx4,cy1,cy2]=[-1<=x,x<0,0<=x,x<=1,x^2+y^2<=1,abs(y)<=1-x]
[bx1,bx2,by1,by2]=[(x,-1,0),(x,0,1),(y,-1,1),(y,-1,1)]
(x0,y0)=(2*random()-1,2*random()-1)
cond=(bool(cx1(x=x0))&bool(cx2(x=x0))&bool(cy1(x=x0,y=y0)))|\
(bool(cx3(x=x0))&bool(cx4(x=x0))&bool(cy2(x=x0,y=y0)))
if cond: s='$\in$'
else: s='$\\notin$'
(region_plot(cy1,bx1,by1,alpha=.7)+region_plot(cy2,bx2,by2,alpha=.7)+
point((x0,y0),size=50,color='red')).show(title='(x0,y0)=(%.4f,%.4f) '%(x0,y0)+s)
```
