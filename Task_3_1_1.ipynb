N21 Задана матрица А с 2 столбцами и 60 строками. <br/>
1-й элемент каждой строки представляет абсциссу, а 2-й - ординату одной из 60 точек на плоскости.<br/>
Определить номера тех строчек, которые удалены от начала координат более, чем на заданное расстояние Н.

```python
def check_distance(x,y,h):
    if x^2+y^2>h^2: return True
    else: return False
def create_array(n): return [round(randint(-100,100)/100,3) for i in range(n)]
H=.7; N=60; A=random_matrix(ZZ,N,2,x=-100,y=100)/100
AH=[check_distance(A[i,0],A[i,1],H) for i in range(N)]
AHI=' '.join([str(i+1) for i in range(N) if AH[i]==True]); print('A:%s'%AHI)
X,Y=create_array(N),create_array(N); XYH=[check_distance(X[i],Y[i],H) for i in range(N)]
XYHI=' '.join([str(i+1) for i in range(N) if XYH[i]==True]); print('X&Y:%s'%XYHI)
pa=sum([list_plot([A[i]],title='matrix A',color=hue(int(AH[i])/1.5)) for i in range(N)])
pa+=sum([text(str(AH[i])[0],(A[i,0],A[i,1]),fontsize=12,color=hue(int(AH[i])/1.5)) for i in range(N)])
pxy=sum([list_plot([(X[i],Y[i])],title='arrays X&Y',color=hue(int(XYH[i])/1.5)) for i in range(N)])
pxy+=sum([text(str(XYH[i])[0],(X[i],Y[i]),fontsize=12,color=hue(int(XYH[i])/1.5)) for i in range(N)])
pc=circle((0,0),H,fill=True,alpha=.4)
graphics_array([pa+pc,pxy+pc]).show(figsize=7,gridlines=True)
```

```python
@interact
def _(x=(-1,1,.1),y=(-1,1,.1),h=(.1,.9,.1)):
    ch=check_distance(x,y,h); col=hue(int(ch)/1.5)
    p=list_plot([(x,y)],color=col,title='[(x=%.1f,y=%.1f),h=%.1f]'%(x,y,h))
    p+=circle((0,0),h,fill=True,alpha=.4)+text(str(ch),(x,y),color=col)
    p.show(figsize=7,gridlines=True,ymin=-1,ymax=1,xmin=-1,xmax=1)
```
