# Function Tabulating with a Choice of a Formula
Create an algorithm and a program

```python
def create_table(f,a,x1,dx,n):
    t=[['n','X','Y','Y≈']]
    for i in [1..n]:
        y=f(a=a,x=x1); t.append([i,x1,y,y.n()]); x1+=dx
    return(table(t))
var('a,x')
f1_1=piecewise([((-oo,0),a/2*(exp(x/a)+exp(-x/a))),([0,0],a/2*(exp(x/a)+exp(-x/a))),
               ((0,+oo),4*a^3*(x^2+4*a^2))],var=x)
A1=1; X1,dX1=-2*A1,A1/5 
pretty_print(latex(f1_1)); create_table(f1_1,A1,X1,dX1,12)
```

```python
def f1_2(a,x):
    if x<=0: return a/2*(exp(x/a)+exp(-x/a))
    else: return 4*a^3*(x^2+4*a^2) 
var('a,x'); st1=(latex(a/2*(exp(x/a)+exp(-x/a))),latex(4*a^3*(x^2+4*a^2)))
s1='$x<=0 => f(x)=%s; x>0 => f(x)=%s$'%st1
pretty_print(html(s1)); create_table(f1_2,A1,X1,dX1,12)
```
