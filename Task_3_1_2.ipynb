21. Заданы N конусов радиусом R своего основания и высотой H (всего 2N значений).<br/>
Найти суммарный объем и число конусов, высота которых не превосходит заданной величины Т,<br/>
а радиус основания R больше $\sqrt T$. Объем конуса $V=\frac{\pi R^2 H}{3}$.

```python
from sage.plot.plot3d.shapes import Cone
def check_cone(r,h,t):
    if (t>=h) and (sqrt(t)<r): return True
    else: return False
def create_array(n): return [round(randint(10,100)/100,3) for i in range(n)]
T=.5; N=10; A=random_matrix(ZZ,N,2,x=10,y=100)/100
R,H=create_array(N),create_array(N); RHT=[check_cone(R[i],H[i],T) for i in range(N)]
AT,ATI,AN,AV=[],'',0,0; RHT,RHTI,RHN,RHV=[],'',0,0
for i in range(N):
    AT+=[check_cone(A[i,0],A[i,1],T)]; RHT+=[check_cone(R[i],H[i],T)]
    if AT[i]==True: 
        AN+=1; ATI+=str(i+1)+' '; AV+=pi*A[i,0]^2*A[i,1]/3
    if RHT[i]==True: 
        RHN+=1; RHTI+=str(i+1)+' '; RHV+=pi*R[i]^2*H[i]/3
print('A: %s; N=%s; V=%s'%(ATI,AN,AV))
print('R&H: %s; N=%s; V=%s'%(RHTI,RHN,RHV))
ps=sphere((0,0,0),T,color='magenta',alpha=.2)
pa=sum([Cone(A[i,0],A[i,1],color=hue(int(AT[i])/1.5),alpha=.3).translate(0,0,0) for i in range(N)])
prh=sum([Cone(R[i],H[i],color=hue(int(RHT[i])/1.5),alpha=.3).translate(0,0,0) for i in range(N)])
(pa+ps).show(frame=false,aspect_ratio=1); (prh+ps).show(frame=false,aspect_ratio=1)
```
